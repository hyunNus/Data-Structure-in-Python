## Queue 큐 ##

```python
class Queue:
    def __init__(self):
        self.items = []  # 데이터 저장을 위한 리스트
        self.front_idx = 0  # 다음 dequeue될 값의 인덱스

    def __len__(self):
        # 실제 큐의 길이는 전체 아이템 개수에서 front 인덱스를 뺀 값
        return len(self.items) - self.front_idx

    def enqueue(self, key):
        self.items.append(key)

    def dequeue(self):
        # front_idx가 리스트의 현재 길이와 같아지면,
        # 모든 아이템이 빠져나간 것으로 간주
        if self.front_idx == len(self.items):
            print("Queue is empty")
            return None
        else:
            # front_idx 위치의 아이템을 반환하고, 인덱스를 1 증가
            x = self.items[self.front_idx]
            self.front_idx += 1
            return x

    def front(self):
        # dequeue 코드와 동일 (단, front_index를 1 증가하는 코드만 삭제)
        if self.front_idx == len(self.items):  # 이 기준이 왜 동작?
            print("Queue is empty")
            return None
        else:  # dequeue할 아이템이 없음을 의미
            return self.items[self.front_idx]


# 요세푸스 문제 해결 함수
def Josephus(n, k):
    """
    n: 전체 사람 수
    k: 제거할 순서
    """
    Q = Queue()
    # 1번부터 n번까지의 사람을 큐에 추가
    for v in range(1, n + 1):
        Q.enqueue(v)

    # 큐에 한 명이 남을 때까지 반복 (len(Q) 사용)
    while len(Q) > 1:
        # k-1번째 사람까지는 큐의 맨 뒤로 보냄
        for i in range(k - 1):  # k-1번 째 까지는 뒤로 보내서 살려야 하니까
            Q.enqueue(Q.dequeue())

        # k번째 사람을 큐에서 제거
        Q.dequeue()

    # 마지막으로 남은 한 명을 반환
    return Q.dequeue()


# --- 실행 예시 ---
if __name__ == "__main__":
    n = 7
    k = 3
    survivor = Josephus(n, k)
    print(f"{n}명의 사람 중 {k}번째마다 제거할 때, 마지막 생존자는 {survivor}번입니다.")
```
