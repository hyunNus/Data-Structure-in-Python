````python
class Heap:
    def __init__(self, L=[]):  # default: ë¹ˆ ë¦¬ìŠ¤íŠ¸
        self.A = L
        self.make_heap()  # Aì˜ ê°’ì„ í™ ì„±ì§ˆì´ ë§Œì¡±ë˜ë„ë¡
        # make_heap í•¨ìˆ˜ í˜¸ì¶œ (ì¶”í›„ ì„¤ëª…)

    def __str__(self):
        return str(self.A)
````

````python
# heapify down implementation in Python -> A[k]ë¥¼ ì•„ë˜ë¡œ ë‚´ë ¤ê°€ë©´ì„œ max-heap ì„±ì§ˆì„ ë§Œì¡±í•˜ëŠ” ê³³ìœ¼ë¡œ ì´ë™
def heapify_down(self, k, n):
    # n = í™ì˜ ì „ì²´ ë…¸ë“œìˆ˜ [heap_sortë¥¼ ìœ„í•´ í•„ìš”í•¨]
    # A[k]ë¥¼ í™ ì„±ì§ˆì„ ë§Œì¡±í•˜ëŠ” ìœ„ì¹˜ë¡œ ë‚´ë ¤ê°€ë©´ì„œ ì¬ë°°ì¹˜

    # [â“ ì§ˆë¬¸ 1]
    while 2 * k + 1 < n:  # kì˜ ì™¼ìª½ ìì‹ ë…¸ë“œê°€ í™ í¬ê¸°(n) ì•ˆì— ìˆëŠ”ì§€ í™•ì¸
        L, R = 2 * k + 1, 2 * k + 2  # L, Rì€ ì™¼ìª½, ì˜¤ë¥¸ìª½ ìì‹ë…¸ë“œ

        # ì„¸ ë…¸ë“œ(ë¶€ëª¨, ì™¼ìª½ìì‹, ì˜¤ë¥¸ìª½ìì‹) ì¤‘ ìµœëŒ€ê°’ì˜ ì¸ë±ìŠ¤ ì°¾ê¸°
        if self.A[L] > self.A[k]:
            m = L
        else:
            m = k

        # ì˜¤ë¥¸ìª½ ìì‹ì´ í™ í¬ê¸°(n) ì•ˆì— ìˆê³ , í˜„ì¬ ìµœëŒ€ê°’(m)ë³´ë‹¤ í¬ë‹¤ë©´
        if R < n and self.A[R] > self.A[m]:
            m = R

        # m = A[k], A[L], A[R] ì¤‘ ìµœëŒ€ê°’ì˜ ì¸ë±ìŠ¤

        if m != k:  # A[k]ê°€ ìµœëŒ€ê°’ì´ ì•„ë‹ˆë©´ (í™ ì„±ì§ˆ ìœ„ë°°)
            self.A[k], self.A[m] = self.A[m], self.A[k]  # ìì‹ê³¼ ê°’ êµí™˜
            k = m  # ì¸ë±ìŠ¤ë¥¼ ìì‹ ë…¸ë“œë¡œ ì´ë™í•˜ì—¬ ê³„ì† ì§„í–‰
        else:
            # [â“ ì§ˆë¬¸ 2]
            break  # A[k]ê°€ ì´ë¯¸ ìì‹ë“¤ë³´ë‹¤ í¬ê±°ë‚˜ ê°™ìœ¼ë¯€ë¡œ (í™ ì„±ì§ˆ ë§Œì¡±) ì¢…ë£Œ
````

````python
# make heap
def make_heap(self):  # nlogn -> O(n)
    n = len(self.A)
    for k in range(n - 1, -1, -1):  # A[n-1] â†’ A[0] ìˆœì„œë¡œ self.heapify_down(k, n)
        self.heapify_down(k, n)  # O(log n)
````

````python
def find_max(self):
    if len(self.A) == 0:
        return None
    return self.A[0]
````

````python
class Heap:
    def __init__(self):
        self.A = []
        self.n = 0  # í™ì˜ ì‹¤ì œ í¬ê¸°ë¥¼ ê´€ë¦¬í•˜ëŠ” ë³€ìˆ˜ ì¶”ê°€

    def delete_max(self):  # O(log n)
        if self.n == 0:  # ë¹ˆ í™ì¸ ê²½ìš°
            return None
        max_value = self.A[0]
        self.A[0] = self.A[self.n - 1]  # ë§ˆì§€ë§‰ ì›ì†Œë¥¼ ë£¨íŠ¸ë¡œ ì´ë™
        self.n -= 1  # í™ í¬ê¸°ë¥¼ 1 ê°ì†Œ
        self.heapify_down(0, self.n)  # í™ ì„±ì§ˆ ë³µì› (ì£¼ì˜: pop()ëŒ€ì‹  nì„ ì‚¬ìš©)
        return max_value
````

````python
def heap_sort(self):
    n = len(self.A)
    self.make_heap()  # í™ ìƒì„±
    for k in range(n - 1, 0, -1):
        self.A[0], self.A[k] = self.A[k], self.A[0]  # ìµœëŒ€ê°’ì„ ë§¨ ë’¤ë¡œ ì´ë™
        self.heapify_down(0, k)  # í™ í¬ê¸°ë¥¼ kë¡œ ì¤„ì—¬ì„œ í™ ì„±ì§ˆ ë³µì›
````

# A[k]ë¥¼ í™ ì„±ì§ˆì„ ë§Œì¡±í•˜ëŠ” ìœ„ì¹˜ë¡œ ì˜¬ë¼ê°€ë©´ì„œ ì¬ë°°ì¹˜
def heapify_up(self, k):

    while k > 0 and self.A[(k - 1) // 2] < self.A[k]:

### 3\. ğŸš€ ì—°ì‚° ë° ìˆ˜í–‰ ì‹œê°„ ì •ë¦¬

(nê°œì˜ ê°’ì„ ì €ì¥í•œ í™ ê¸°ì¤€)

  * `heapify_up`: **$O(log n)$**
  * `heapify_down`: **$O(log n)$**
  * `insert`: (1 x `heapify_up`) = **$O(log n)$**
  * `delete_max`: (1 x `heapify_down`) = **$O(log n)$**
  * `make_heap`: (n/2ë²ˆì˜ `heapify_down` í˜¸ì¶œ) = **$O(n)$**
      * *ì°¸ê³ : ë‹¨ìˆœ ê³„ì‚°ìœ¼ë¡œëŠ” $O(n log n)$ì²˜ëŸ¼ ë³´ì´ì§€ë§Œ, ë” íƒ€ì´íŠ¸í•œ ë¶„ì„(tight bound)ì„ í†µí•´ $O(n)$ì„ì´ ì¦ëª…ë©ë‹ˆë‹¤.*
  * `heap_sort`: (`make_heap` + n x `delete_max`) = $O(n) + O(n log n)$ = **$O(n log n)$**

- insert(x) O(logn)

- decrease_key

- increase key

````python
import heapq  # min heap


# max hip í•˜ê³  ì‹¶ìœ¼ë©´ -1 ê³±í•´ì„œ ê¸°ë¦¿
class MaxHeap:
    def __init__(self):
        self.heap = []

    def insert(self, val):
        heapq.heappush(self.heap, -val)  # ìŒìˆ˜ë¡œ ì €ì¥í•˜ì—¬ ìµœëŒ€ í™ êµ¬í˜„

    def delete_max(self):
        if not self.heap:
            return None
        return -heapq.heappop(self.heap)  # ìŒìˆ˜ë¥¼ ë‹¤ì‹œ ì–‘ìˆ˜ë¡œ ë³€í™˜í•˜ì—¬ ë°˜í™˜
````
