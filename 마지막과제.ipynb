# 1번 문제


```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


class BST:
    def __init__(self):
        self.root = None
        self.size = 0

    def preorder(self, node):
        if node:
            print(node.key, end=" ")
            self.preorder(node.left)
            self.preorder(node.right)


def build_bst(post_order, in_order, post_start, post_end, in_start, in_end):
    if post_start > post_end or in_start > in_end:
        return None

    root_val = post_order[post_end]
    node = Node(root_val)

    root_idx = in_order.index(root_val)
    left_count = root_idx - in_start

    node.left = build_bst(
        post_order,
        in_order,
        post_start,
        post_start + left_count - 1,
        in_start,
        root_idx - 1,
    )
    node.right = build_bst(
        post_order,
        in_order,
        post_start + left_count,
        post_end - 1,
        root_idx + 1,
        in_end,
    )

    return node


post_order = [int(x) for x in input().split()]
n = len(post_order)
T = BST()
T.root = build_bst(post_order, sorted(post_order), 0, n - 1, 0, n - 1)
T.size = n

T.preorder(T.root)
```

# 2번 문제 

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.parent = self.left = self.right = None
        self.height = 0

    def __str__(self):
        return str(self.key)


class BST:
    def __init__(self):
        self.root = None
        self.size = 0

    def __len__(self):
        return self.size

    def preorder(self, v):
        if v != None:
            print(v.key, end=" ")
            self.preorder(v.left)
            self.preorder(v.right)

    def inorder(self, v):
        if v != None:
            self.inorder(v.left)
            print(v.key, end=" ")
            self.inorder(v.right)

    def postorder(self, v):
        if v != None:
            self.postorder(v.left)
            self.postorder(v.right)
            print(v.key, end=" ")

    def find_loc(self, key):
        if self.size == 0:
            return None
        p = None
        v = self.root
        while v:
            if v.key == key:
                return v
            p = v
            if v.key < key:
                v = v.right
            else:
                v = v.left
        return p

    def search(self, key):
        v = self.root
        while v:
            if v.key == key:
                return v
            elif v.key < key:
                v = v.right
            else:
                v = v.left
        return None

    def insert(self, key):
        p = self.find_loc(key)
        if p and p.key == key:
            return p

        new_node = Node(key)
        if p is None:
            self.root = new_node
        else:
            new_node.parent = p
            if p.key < key:
                p.right = new_node
            else:
                p.left = new_node

        self.size += 1
        self.update_height(new_node)
        return new_node

    def deleteByMerging(self, x):
        if x is None:
            return
        a, b, pt = x.left, x.right, x.parent
        s = pt

        if a is None and b is None:
            if pt:
                if pt.left == x:
                    pt.left = None
                else:
                    pt.right = None
            else:
                self.root = None

        elif a is None:
            c = b
            if pt:
                if pt.left == x:
                    pt.left = c
                else:
                    pt.right = c
            else:
                self.root = c
            c.parent = pt
            s = pt

        elif b is None:
            c = a
            if pt:
                if pt.left == x:
                    pt.left = c
                else:
                    pt.right = c
            else:
                self.root = c
            c.parent = pt
            s = pt

        else:
            m = a
            while m.right:
                m = m.right
            m.right = b
            b.parent = m

            if pt:
                if pt.left == x:
                    pt.left = a
                else:
                    pt.right = a
            else:
                self.root = a
            a.parent = pt
            s = m

        self.size -= 1
        self.update_height(s)

    def deleteByCopying(self, x):
        if x is None:
            return

        node_to_delete = x

        if x.left and x.right:
            node_to_delete = self.pred(x)
            x.key = node_to_delete.key

        child = node_to_delete.left if node_to_delete.left else node_to_delete.right
        pt = node_to_delete.parent

        if child:
            child.parent = pt

        if pt:
            if pt.left == node_to_delete:
                pt.left = child
            else:
                pt.right = child
        else:
            self.root = child

        self.size -= 1
        self.update_height(pt)

    def height(self, x):
        if x == None:
            return -1
        else:
            return x.height

    def update_height(self, v):
        while v:
            new_h = 1 + max(self.height(v.left), self.height(v.right))
            v.height = new_h
            v = v.parent

    def succ(self, x):
        if x is None:
            return None
        if x.right:
            v = x.right
            while v.left:
                v = v.left
            return v
        else:
            p = x.parent
            while p and p.right == x:
                x = p
                p = p.parent
            return p

    def pred(self, x):
        if x is None:
            return None
        if x.left:
            v = x.left
            while v.right:
                v = v.right
            return v
        else:
            p = x.parent
            while p and p.left == x:
                x = p
                p = p.parent
            return p

    def rotateLeft(self, x):
        if x.right is None:
            return
        z = x.right

        b = z.left
        z.parent = x.parent
        if x.parent:
            if x.parent.left == x:
                x.parent.left = z
            else:
                x.parent.right = z
        else:
            self.root = z

        z.left = x
        x.parent = z
        x.right = b
        if b:
            b.parent = x

        self.update_height(x)

    def rotateRight(self, x):
        if x.left is None:
            return
        z = x.left

        b = z.right
        z.parent = x.parent
        if x.parent:
            if x.parent.left == x:
                x.parent.left = z
            else:
                x.parent.right = z
        else:
            self.root = z

        z.right = x
        x.parent = z
        x.left = b
        if b:
            b.parent = x

        self.update_height(x)


def build_bst(post_order, in_order, post_start, post_end, in_start, in_end):
    if post_start > post_end or in_start > in_end:
        return None

    root_val = post_order[post_end]
    node = Node(root_val)

    root_idx = in_order.index(root_val)
    left_count = root_idx - in_start

    node.left = build_bst(
        post_order,
        in_order,
        post_start,
        post_start + left_count - 1,
        in_start,
        root_idx - 1,
    )
    node.right = build_bst(
        post_order,
        in_order,
        post_start + left_count,
        post_end - 1,
        root_idx + 1,
        in_end,
    )

    h_left = -1
    h_right = -1

    if node.left:
        node.left.parent = node
        h_left = node.left.height

    if node.right:
        node.right.parent = node
        h_right = node.right.height

    node.height = max(h_left, h_right) + 1

    return node


if __name__ == "__main__":

    post_order = [int(x) for x in input().split()]
    n = len(post_order)
    T = BST()
    T.root = build_bst(post_order, sorted(post_order), 0, n - 1, 0, n - 1)
    T.size = n

    while True:
        cmd = input().split()
        if cmd[0] == "insert":
            v = T.insert(int(cmd[1]))
            print(f"+ {v.key} is inserted")
        elif cmd[0] == "deleteC":
            v = T.search(int(cmd[1]))
            T.deleteByCopying(v)
            print(f"- {int(cmd[1])} is deleted by copying")
        elif cmd[0] == "deleteM":
            v = T.search(int(cmd[1]))
            T.deleteByMerging(v)
            print(f"- {int(cmd[1])} is deleted by merging")
        elif cmd[0] == "search":
            v = T.search(int(cmd[1]))
            if v == None:
                print(f"* {cmd[1]} is not found!")
            else:
                print(f"* {cmd[1]} is found!")
        elif cmd[0] == "height":
            h = T.height(T.search(int(cmd[1])))
            if h == -1:
                print(f"= {cmd[1]} is not found!")
            else:
                print(f"= {cmd[1]} has height of {h}")
        elif cmd[0] == "succ":
            v = T.succ(T.search(int(cmd[1])))
            if v == None:
                print(f"> {cmd[1]} is not found or has no successor")
            else:
                print(f"> {cmd[1]}'s successor is {v.key}")
        elif cmd[0] == "pred":
            v = T.pred(T.search(int(cmd[1])))
            if v == None:
                print(f"< {cmd[1]} is not found or has no predecssor")
            else:
                print(f"< {cmd[1]}'s predecssor is {v.key}")
        elif cmd[0] == "Rleft":
            v = T.search(int(cmd[1]))
            if v == None:
                print(f"@ {cmd[1]} is not found!")
            else:
                T.rotateLeft(v)
                print(f"@ Rotated left at node {cmd[1]}")
        elif cmd[0] == "Rright":
            v = T.search(int(cmd[1]))
            if v == None:
                print(f"@ {cmd[1]} is not found!")
            else:
                T.rotateRight(v)
                print(f"@ Rotated right at node {cmd[1]}")
        elif cmd[0] == "preorder":
            T.preorder(T.root)
            print()
        elif cmd[0] == "postorder":
            T.postorder(T.root)
            print()
        elif cmd[0] == "inorder":
            T.inorder(T.root)
            print()
        elif cmd[0] == "exit":
            break
        else:
            print("* not allowed command. enter a proper command!")
```

# 3번 문제

```python
class Node:
    def __init__(self, key, parent=None):
        self.key = key
        self.left = None
        self.right = None
        self.parent = parent
        self.height = 0  # leaf = 0


class BST:
    def __init__(self):
        self.root = None
        self.size = 0

    def height(self, node):
        return node.height if node else -1

    def search(self, key):
        t = self.root
        while t:
            if key == t.key:
                return t
            elif key < t.key:
                t = t.left
            else:
                t = t.right
        return None

    def insert(self, key):
        if self.root is None:
            self.root = Node(key)
            self.size = 1
            return self.root

        p = None
        t = self.root
        while t:
            if key == t.key:
                return t
            p = t
            if key < t.key:
                t = t.left
            else:
                t = t.right

        new_node = Node(key, p)

        if key < p.key:
            p.left = new_node
        else:
            p.right = new_node

        v = new_node
        while v:
            v.height = 1 + max(self.height(v.left), self.height(v.right))
            v = v.parent

        self.size += 1
        return new_node

    def deleteByCopying(self, u):
        if u is None:
            return None

        if u.left is None or u.right is None:
            child = u.left if u.left else u.right
            parent = u.parent

            if parent is None:
                self.root = child
                if child:
                    child.parent = None
            else:
                if parent.left == u:
                    parent.left = child
                else:
                    parent.right = child
                if child:
                    child.parent = parent

            self.size -= 1
            return parent

        else:
            pred = u.left
            while pred.right:
                pred = pred.right

            u.key = pred.key
            parent = pred.parent
            child = pred.left

            if parent.left == pred:
                parent.left = child
            else:
                parent.right = child

            if child:
                child.parent = parent

            self.size -= 1
            return parent

    def preorder(self, node):
        if node:
            print(node.key, end="")
            self.preorder(node.left)
            self.preorder(node.right)

    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.key, end="")
            self.inorder(node.right)

    def postorder(self, node):
        if node:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.key, end="")

    def succ(self, node):
        if node is None:
            return None
        if node.right:
            t = node.right
            while t.left:
                t = t.left
            return t

        t = node
        while t.parent and t == t.parent.right:
            t = t.parent
        return t.parent

    def pred(self, node):
        if node is None:
            return None
        if node.left:
            t = node.left
            while t.right:
                t = t.right
            return t

        t = node
        while t.parent and t == t.parent.left:
            t = t.parent
        return t.parent


class AVL(BST):
    def __init__(self):
        self.root = None
        self.size = 0

    def update_height(self, v):
        if v:
            v.height = 1 + max(self.height(v.left), self.height(v.right))

    def rotate_left(self, z):
        y = z.right
        y.parent = z.parent

        z.right = y.left
        if y.left:
            y.left.parent = z

        y.left = z
        z.parent = y

        if y.parent:
            if y.parent.left == z:
                y.parent.left = y
            else:
                y.parent.right = y
        else:
            self.root = y

        self.update_height(z)
        self.update_height(y)
        return y

    def rotate_right(self, z):
        y = z.left
        y.parent = z.parent

        z.left = y.right
        if y.right:
            y.right.parent = z

        y.right = z
        z.parent = y

        if y.parent:
            if y.parent.left == z:
                y.parent.left = y
            else:
                y.parent.right = y
        else:
            self.root = y

        self.update_height(z)
        self.update_height(y)
        return y

    def rebalance(self, x, y, z):
        # LL
        if z.left == y and y.left == x:
            return self.rotate_right(z)

        # LR
        if z.left == y and y.right == x:
            self.rotate_left(y)
            return self.rotate_right(z)

        # RR
        if z.right == y and y.right == x:
            return self.rotate_left(z)

        # RL
        if z.right == y and y.left == x:
            self.rotate_right(y)
            return self.rotate_left(z)

        return z

    def insert(self, key):
        v = super(AVL, self).insert(key)
        t = v.parent

        while t:
            self.update_height(t)
            balance = self.height(t.left) - self.height(t.right)

            if abs(balance) > 1:
                z = t
                y = z.left if self.height(z.left) >= self.height(z.right) else z.right
                x = y.left if self.height(y.left) >= self.height(y.right) else y.right

                new_top = self.rebalance(x, y, z)

                p = new_top.parent
                while p:
                    self.update_height(p)
                    p = p.parent
                break
            t = t.parent

        return v

    def delete(self, u):
        v = self.deleteByCopying(u)

        t = v
        while t:
            self.update_height(t)
            balance = self.height(t.left) - self.height(t.right)

            if abs(balance) > 1:
                z = t
                y = z.left if self.height(z.left) >= self.height(z.right) else z.right
                x = y.left if self.height(y.left) >= self.height(y.right) else y.right

                new_top = self.rebalance(x, y, z)

                p = new_top.parent
                while p:
                    self.update_height(p)
                    p = p.parent

                t = new_top.parent
            else:
                t = t.parent


T = AVL()
while True:
    cmd = input().split()
    if cmd[0] == "insert":
        v = T.insert(int(cmd[1]))
        print(f"+ {v.key} is inserted")
    elif cmd[0] == "delete":
        v = T.search(int(cmd[1]))
        T.delete(v)
        print(f"- {int(cmd[1])} is deleted")
    elif cmd[0] == "search":
        v = T.search(int(cmd[1]))
        if v == None:
            print(f"* {cmd[1]} is not found!")
        else:
            print(f"* {cmd[1]} is found!")
    elif cmd[0] == "height":
        h = T.height(T.search(int(cmd[1])))
        if h == -1:
            print(f"= {cmd[1]} is not found!")
        else:
            print(f"= {cmd[1]} has height of {h}")
    elif cmd[0] == "succ":
        v = T.succ(T.search(int(cmd[1])))
        if v == None:
            print(f"> {cmd[1]} is not found or has no successor")
        else:
            print(f"> {cmd[1]}'s successor is {v.key}")
    elif cmd[0] == "pred":
        v = T.pred(T.search(int(cmd[1])))
        if v == None:
            print(f"< {cmd[1]} is not found or has no predecssor")
        else:
            print(f"< {cmd[1]}'s predecssor is {v.key}")
    elif cmd[0] == "preorder":
        T.preorder(T.root)
        print()
    elif cmd[0] == "postorder":
        T.postorder(T.root)
        print()
    elif cmd[0] == "inorder":
        T.inorder(T.root)
        print()
    elif cmd[0] == "exit":
        break
    else:
        print("* not allowed command. enter a proper command!")
```
